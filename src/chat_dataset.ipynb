```python
from datasets import load_dataset

ds = load_dataset(
    "csv",
    data_files="hf://datasets/strova-ai/customer_support_conversations_dataset/customer_support_data.csv",
)["train"]

rename_map = {
    "conv_id": "conversation_id",
    "turn_index": "turn_id",
    "text": "message",
    "product": "category",
    "issue_type": "sub_category",
    "language": "locale",
    "customer_name": "customer_id",   # name only if you want compatibility
    "agent_name": "agent_id",         # name only if you want compatibility
    "overall_sentiment": "sentiment",
    "overall_urgency": "priority",
    "outcome": "status",
    "primary_intent": "intent",
}

for old, new in rename_map.items():
    if old in ds.column_names:
        ds = ds.rename_column(old, new)

print(ds.column_names)
print(ds[0])
```
