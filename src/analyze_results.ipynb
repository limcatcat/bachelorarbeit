```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = [
    "../comparisons/tables/AF_boolq_gpt-4o_comparison.csv",
    "../comparisons/tables/RF_boolq_gpt-4o_comparison.csv",
    "../comparisons/tables/NR_boolq_gpt-4o_comparison.csv",
]

dfs_boolq = [pd.read_csv(f) for f in files]
comparison_df_boolq = pd.concat(dfs_boolq, ignore_index=True)

accuracy_by_variant_boolq = comparison_df_boolq.groupby("variant")["answer_match"].mean()
print(accuracy_by_variant_boolq)
```

```python
accuracy_by_variant_boolq = comparison_df_boolq.groupby("variant")["answer_match"].mean()

accuracy_by_variant_boolq.plot(kind="bar", color="#5ab8a2")
plt.suptitle("BoolQ with gpt-4o", fontsize=10)
plt.title("Accuracy by Prompt Variant")
plt.ylabel("Accuracy")
plt.xlabel("Variant")
plt.xticks(rotation=0)
plt.show()
```

```python
files = [
    # gpt-4o
    "../comparisons/tables/AF_boolq_gpt-4o_comparison.csv",
    "../comparisons/tables/RF_boolq_gpt-4o_comparison.csv",
    "../comparisons/tables/NR_boolq_gpt-4o_comparison.csv",

    # gpt-5.4
    "../comparisons/tables/AF_boolq_gpt-5.4_comparison.csv",
    "../comparisons/tables/RF_boolq_gpt-5.4_comparison.csv",
    "../comparisons/tables/NR_boolq_gpt-5.4_comparison.csv",

    # o3
    "../comparisons/tables/AF_boolq_o3_comparison.csv",
    "../comparisons/tables/RF_boolq_o3_comparison.csv",
    "../comparisons/tables/NR_boolq_o3_comparison.csv",
]

dfs = [pd.read_csv(f) for f in files]
comparison_df = pd.concat(dfs, ignore_index=True)

pivot = comparison_df.pivot_table(
    index="variant",
    columns="model_name",
    values="answer_match",
    aggfunc="mean"
)

print(pivot)
```

```python
import matplotlib.pyplot as plt

pivot.plot(kind="bar", color=["#faa30c", "#600b99", "#6E4326"],)

plt.title("BoolQ: Accuracy by Variant and Model")
plt.ylabel("Accuracy")
plt.xlabel("Variant")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend(title="Model", loc="lower right")
plt.show()
```
